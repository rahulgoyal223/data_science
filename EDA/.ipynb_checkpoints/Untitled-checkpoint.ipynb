```python
bating=pd.read_csv("odi-batting.csv")
bating.head()
```

```python
bating["century"] = bating.Runs.apply(lambda x: 1 if x>=100 else 0)
```

```python
bating.head()
```

```python
bating.pivot_table(values="century", index="Player", aggfunc=np.sum).sort_values(by="century",ascending=False)
```

```python
bating["strike"] = round((bating.Runs/bating.Balls)*100,0).astype(int)
```

```python
bating.head()
```

```python
bat=bating.loc[bating.century==1]
```

```python
bat[bat.strike==bat.strike.max()]
```

```python
bating["year"] = bating["MatchDate"].apply(lambda x: x.split("-")[2])
```

```python
bating.loc[bating.Country=="India"].pivot_table(values="century", index="year", aggfunc=np.sum).sort_values(by="century",ascending=False)
```
