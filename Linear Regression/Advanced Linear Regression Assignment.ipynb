## Null Treatment
- GarageYrBlt - This column is not of much importance. Hence,It can be deleted.
- MSSubClass - The imformation represent by this column also represented by age and HouseStyle. Hence Dropping
- LotFrontage - This column has 17 percent null values. And we have just 1460 records. Hence, droppeing the column.
- MasVnrArea - Droping rows with null values. (.5 percent). Imputing with 0.0
- MasVnrType - Imputing None
- Electrical - Imputing with Mode
- For below columns imputing 'NA' in place of Null.
 1. PoolQC
 2. MiscFeature
 3. Alley
 4. Fence
 5. FireplaceQu
 6. GarageCond
 7. GarageType
 8. GarageFinish
 9. GarageQual
 10. BsmtFinType2
 11. BsmtExposure
 12. BsmtQual
 13. BsmtCond
 14. BsmtFinType1

```python
#Dropping GarageYrBlt column
housing = housing.drop(['GarageYrBlt','LotFrontage','MSSubClass'],axis=1)
housing.shape
```

```python
# Imputing 'None' with Null for MasVnrType
housing['MasVnrType'] = housing['MasVnrType'].fillna('None')
```

```python
# Imputing Mode with Null for Electrical
housing['Electrical'] = housing['Electrical'].fillna(list(housing['Electrical'].mode())[0])
```

```python
#Imputing pending columns with NA
housing = housing.fillna('NA')
```

```python
import statsmodels.api as sm
def getModel(df):
    # Add a constant
    X_train_lm = sm.add_constant(df)
    #Training model
    lr = sm.OLS(y_train, X_train_lm).fit()
    print(lr.summary())
    return lr, X_train_lm
```

```python
# Method to calculate VIF.
from statsmodels.stats.outliers_influence import variance_inflation_factor
def cal_vif(df):
    vif = pd.DataFrame()
    X = df
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```
