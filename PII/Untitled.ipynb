```python
#code to get latest version
def getlatestVersion(command):
    os.system('{} > tmp'.format(command))
    f =  open('tmp', 'r').read()
    y = json.loads(f)
    return y['Versions'][0]['VersionId']
```

```python
def set_payload(fileName,s3InputVideoPath,deviceId,txnHash,requestId,key):
    sample_json = '''{
"numberOfMetadataFiles": 0,
  "traceId": "X@wgU@ptuDUehwpL4h-4VAAAAAk",
  "fileName": "0_trip1_part1_53.4607_-2.3359_0.0_1608106841493_y.mp4",
  "videoDetailsRecordId": 0,
  "partId": 1,
  "s3InputVideoPath": "https://fleetdata-protected-production.s3.amazonaws.com/N641749401571071/74da10855034a60de7942ddfdd345e7e57b8ec6f/trimmedVideos/alert/1_1320_1799/inputVideo.mp4",
  "alreadyAvailable": true,
  "deviceId": "16318410",
  "blurring_params": {
    "blur_objects_list": [
      "face",
      "license_plate"
    ]
  },
  "txnHash": "6cbdd0b418a04ef01d2b3f66e8f88aeda93f8531",
  "requestId": 14444207,
  "function": "blur",
  "tenantId": 3475,
  "tenantBucket": "fleetdata-production",
  "key": "N641749401571071/74da10855034a60de7942ddfdd345e7e57b8ec6f"
}'''
    payload = json.loads(sample_json)
    payload['fileName'] = fileName
    payload['s3InputVideoPath'] = s3InputVideoPath.replace('fleetdata-production','fleetdata-protected-production')
    payload['deviceId'] = deviceId
    payload['txnHash'] = txnHash
    payload['requestId'] = requestId
    payload['key'] = key
    return payload
```
